```python
#Importar dados essenciais
import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive/')

dataset = pd.read_csv('/content/drive/My Drive/CursoIA/Multiclass Classification  - base_iris/iris.csv')
dataset
```

# Separando as partes de previsores e classes

```python
previsores = dataset.iloc[:, 0:4].values
previsores
```

```python
classe = dataset.iloc[:, 4].values
classe
```

```python
#Importando lib para split da base
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder #Para transformar classes em nÃºmeros
from tensorflow.keras.utils import to_categorical #Para criar formatar classes com formato 100, 010, 001, de acordo com sua classe


labelEncoder = LabelEncoder()
classe =  labelEncoder.fit_transform(classe)
print(classe)
classe_dummy = to_categorical(classe)
print(classe_dummy)
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classe_dummy, test_size=0.25)
```

```python
#Importando Rede neural
from keras.models import Sequential
from keras.layers import Dense

classificador = Sequential()
classificador.add(Dense(units =  4, # (inputs + outputs)/2
                        activation = 'relu',
                        input_dim = 4))
classificador.add(Dense(units =  4,
                        activation = 'relu'))
classificador.add(Dense(units =  3,
                        activation = 'softmax', # Usada para mais 2 classes de saÃ­da
                        input_dim = 4))
classificador.compile(optimizer = 'adam',
                      loss = 'categorical_crossentropy',
                      metrics = ['categorical_accuracy'])
```

#Treinando rede neural

```python
classificador.fit(previsores_treinamento,
                  classe_treinamento,
                  batch_size = 10,
                  epochs = 100)
```

```python
# Verificando resultados
resultados = classificador.evaluate(previsores_teste, classe_teste)
print(resultados)
previsoes = classificador.predict(previsores_teste)
previsoes = (previsoes > 0.5)
classe_teste2 = [np.argmax(t) for t in classe_teste]
previsoes2 = [np.argmax(t) for t in previsoes]
print(previsoes2)
print(classe_teste2)

from sklearn.metrics import confusion_matrix

matrix =  confusion_matrix(previsoes2, classe_teste2)

print(matrix)
```
